```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Data Loading and Preprocessing

```python
train_df = pd.read_csv(r'DATA  Scientist Assignment\train.csv')
test_df = pd.read_csv(r'DATA  Scientist Assignment\test.csv')
trainlabel_df=pd.read_csv(r"DATA  Scientist Assignment\trainLabels.csv")
```

```python
# Align test_df columns with train_df
test_df.columns = train_df.columns
```

```python
categorical_cols = train_df.select_dtypes(include=['object']).columns
```

```python
# Function to identify hash columns
def is_hash_column(series):
    sample_values = series.dropna().astype(str).sample(min(10, len(series)))
    return any(len(val) > 10 and any(char in val for char in '0123456789abcdef') for val in sample_values)
```

```python

# Filter out hash columns
non_hash_categorical_cols = [col for col in categorical_cols if not is_hash_column(train_df[col])]
```

```python
label_encoder = LabelEncoder()
```

```python
# Apply LabelEncoder to each non-hash categorical column in both train and test data
for col in non_hash_categorical_cols:
    train_df[col] = label_encoder.fit_transform(train_df[col].astype(str))
    test_df[col] = label_encoder.transform(test_df[col].astype(str))
```

```python
# Identify hash columns
hash_columns = [col for col in categorical_cols if is_hash_column(train_df[col])]
```

```python
# Function to hash a string using hashlib
import hashlib

def hash_string(value):
    if pd.isnull(value):
        return -1  # A placeholder for missing values
    return int(hashlib.sha256(value.encode('utf-8')).hexdigest(), 16) % (10 ** 8)
```

```python
# Apply the hash function to the hash columns in both train and test data
for col in hash_columns:
    train_df[col] = train_df[col].apply(lambda x: hash_string(str(x)))
    test_df[col] = test_df[col].apply(lambda x: hash_string(str(x)))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
```

```python
train_df = train_df.drop_duplicates().reset_index(drop=True)
trainlabel_df = trainlabel_df.drop_duplicates().reset_index(drop=True)
```

```python
print(train_df.shape,trainlabel_df.shape)
```

```python
# Align indices
if train_df.shape[0] > trainlabel_df.shape[0]:
    train_df = train_df.loc[trainlabel_df.index]
elif train_df.shape[0] < trainlabel_df.shape[0]:
    trainlabel_df = trainlabel_df.loc[train_df.index]

# Ensure no missing rows in aligned dataframes
train_df = train_df.dropna().reset_index(drop=True)
trainlabel_df = trainlabel_df.dropna().reset_index(drop=True)

# Final shape check
print(f"Final Shape of train_df: {train_df.shape}")
print(f"Final Shape of trainlabel_df: {trainlabel_df.shape}")
```

```python
import pandas as pd
import numpy as np
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.ensemble import GradientBoostingClassifier
```

```python
X_train, X_val, y_train, y_val = train_test_split(train_df, trainlabel_df, test_size=0.2, random_state=42)
```

```python
import lightgbm as lgb
```

```python
# Initialize the LightGBMClassifier
lgb_classifier = lgb.LGBMClassifier(n_estimators=50, max_depth=3, random_state=42)
```

```python

# Wrap the classifier with MultiOutputClassifier
multi_target_lgb = MultiOutputClassifier(lgb_classifier, n_jobs=-1)
```

```python
# Fit the model
multi_target_lgb.fit(X_train, y_train)
```

```python
# Predict on the validation set
y_pred = multi_target_lgb.predict(X_val)
```

```python
# Evaluate the model for each target separately
for i, col in enumerate(trainlabel_df.columns):
    print(f"Classification Report for {col}:")
    print(classification_report(y_val.iloc[:, i], y_pred[:, i]))
```

```python
# Predict on the test data
test_predictions = multi_target_lgb.predict(test_df)
```

```python
# Convert predictions to DataFrame
test_predictions_df = pd.DataFrame(test_predictions, columns=trainlabel_df.columns)
```
